### Question 6: Python program using regex to extract email addresses

```python
import re

text = """Hello team, please contact us at support@xyz.com for technical issues,
or reach out to our HR at hr@xyz.com. You can also connect with John at
john.doe@xyz.org and jenny via jenny_clarke126@mail.co.us.
For partnership inquiries, email partners@xyz.biz."""

pattern = r'[a-zA-Z0-9._%+-]+@[a-zA-Z0-9.-]+\.[a-zA-Z]+'

emails = re.findall(pattern, text)

print("Extracted Email Addresses:")
for email in emails:
    print(email)
```

### Question 7: Tokenization and Frequency Distribution using Python and NLTK

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist

text = """Natural Language Processing (NLP) is a fascinating field that combines
linguistics, computer science, and artificial intelligence. It enables machines
to understand, interpret, and generate human language. Applications of NLP
include chatbots, sentiment analysis, and machine translation. As technology
advances, the role of NLP in modern solutions is becoming increasingly critical."""

tokens = word_tokenize(text.lower())

fdist = FreqDist(tokens)

print("Tokens:")
print(tokens)

print("\nWord Frequency Distribution:")
print(fdist.most_common(10))
```

### Question 8: Custom annotator using spaCy to identify proper nouns

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = "Apple CEO Tim Cook visited India and met Prime Minister Narendra Modi."

doc = nlp(text)

print("Proper Nouns Identified:")

for token in doc:
    if token.pos_ == "PROPN":
        print(token.text)
```

### Question 9: Training a Word2Vec model using Gensim

```python
!pip install gensim
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize

dataset = [
"Natural language processing enables computers to understand human language",
"Word embeddings are a type of word representation that allows words with similar meaning to have similar representation",
"Word2Vec is a popular word embedding technique used in many NLP applications",
"Text preprocessing is a critical step before training word embeddings",
"Tokenization and normalization help clean raw text for modeling"
]

tokenized_data = [word_tokenize(sentence.lower()) for sentence in dataset]

model = Word2Vec(sentences=tokenized_data, vector_size=100, window=5, min_count=1, workers=4)

print("Vector for word 'language':")
print(model.wv['language'])
```

### Question 10: NLP pipeline for analyzing customer feedback in a fintech startup
Answer:

To analyze thousands of customer reviews using NLP, the following steps would be followed:

1. **Data Collection:**
Gather customer feedback from various sources such as app reviews, emails, support tickets, and surveys.

2. **Data Cleaning:**
Remove noise from the text by eliminating punctuation, special characters, HTML tags, and duplicate records.

3. **Text Normalization:**
Convert text to lowercase, remove stopwords, and perform stemming or lemmatization.

4. **Tokenization:**
Split the text into individual words or tokens for easier analysis.

5. **Feature Extraction:**
Convert text into numerical representations using techniques such as:

    - Bag of Words

    - TF-IDF

    - Word Embeddings


6. **Sentiment Analysis:**
Use machine learning models to classify feedback into positive, negative, or neutral sentiment.

7. **Topic Modeling:**
Apply algorithms like Latent Dirichlet Allocation (LDA) to identify major topics discussed in customer feedback.

8. **Visualization and Insights:**
Create dashboards or reports to show trends, major complaints, feature requests, and customer satisfaction levels.

These insights help fintech companies improve products, enhance customer experience, and make better business decisions.
